```python
def weight_check(kits, k):
    weight = 500
    for kit in kits:
        weight += (kit-k)
        if weight < 500:
            return False
    return True

n, k = map(int, input().split())
kits = list(map(int, input().split()))

from itertools import permutations
answer = 0
for p in permutations(kits, n):
    if weight_check(p, k): answer += 1
print(answer)
```
