```python
n, m = map(int, input().split())
graph = []
for i in range(n):
    graph.append(list(map(int, input())))
```

```python
# 이동방향 정의(상, 하, 좌, 우)
dx = [0, 0, -1, 1]
dy = [1, -1, 0, 0]
```

```python
from collections import deque
def bfs(x,y):
    queue = deque()
    queue.append((x,y))
    # 큐가 빌때까지 반복
    while queue:
        x, y = queue.popleft()
        # 현재위치에서 네 방향으로 위치 확인
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            # 공간을 벗어나면 무시
            if nx < 0 or ny <0 or nx >= n or ny >= m:
                continue
            # 0인 경우 무시
            if graph[nx][ny] == 0:
                continue
            # 해당노드를 처음 방문하는 경우에만 최단 거리 기록
            if graph[nx][ny] == 1:
                graph[nx][ny] = graph[x][y] + 1
                queue.append((nx, ny))
        #도착지점 반환
    return graph[n-1][m-1]
```

```python
print(bfs(0,0))
```
