```python
n = int(input())
numbers = []
for i in range(n):
    numbers.append(int(input()))
# 리스트를 내림차순으로 정렬하고 출력
for n in sorted(numbers, reverse = True):
    print(n, end=' ')
```

```python
n = int(input())
array = []
for i in range(n):
    array.append(int(input()))

array = sorted(array, reverse = True)
for i in array:
    print(i, end=' ')
```
