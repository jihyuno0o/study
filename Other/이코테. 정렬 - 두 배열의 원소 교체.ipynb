```python
n, k = map(int, input().split())
A = sorted(map(int, input().split()))
B = sorted(map(int, input().split()), reverse = True)
```

```python
for i in range(k):
    if A[i] < B[i]:
        A[i], B[i] = B[i], A[i]
    else: break
```

```python
print(sum(A))
```
