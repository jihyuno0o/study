```python
# 18번 실패
def solution(n):
    if n == 1: return 1
    answer = 1 # 15 = 15 경우 더함
    
    m = n//2 + 1
    s = []
    
    for i in range(m):
        num = m-i # 8,7,6,5,,,
        s.append(num)
        
        if sum(s) == n: 
            answer += 1
        elif sum(s) > n:
            s.remove(max(s))
    return answer 
```

```python
def solution(n):
    answer = 1 # 15 = 15 경우 더함
    
    for i in range(1, n//2+1):
        sum = 0
        for j in range(i,  n//2+2):
            sum += j
            if sum == n:
                answer += 1
                break
            elif sum > n:
                break
                
    return answer
```
