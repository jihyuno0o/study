```python
n, w, l = map(int, input().split()) #  ,  ,   
```

```python
truck = list(map(int, input().split()))
```

```python
truck.reverse()
truck
```

```python
from collections import deque
bridge = deque([0]*w)
```

```python
bridge
```

```python
answer = 0
sum_weight = 0
```

```python
while truck:
    answer += 1
    bridge.append(bridge.popleft())
    
    if bridge[-1] != 0:
        sum_weight -= bridge[-1]
        bridge[-1] = 0
    
    if sum_weight + truck[-1] <= l:
        bridge[-1] = truck.pop()
        sum_weight += bridge[-1]
```

```python
answer + w
```
